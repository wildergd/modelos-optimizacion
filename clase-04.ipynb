```python
from random import randint
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fx = lambda x: 2 * np.sin(5 * x) + 3 * x ** 3 + - 2 * x ** 2 + 3 * x - 5

delta_x = 0.01
x = np.arange(0, 1, delta_x)
```

```python
plt.plot(x, fx(x))
plt.show()
```

```python
fx_dx = lambda x: 10 * np.cos(5 * x) + 9 * x ** 2 + - 4 * x + 3

fx_dx_forward = lambda x, delta: (fx(x + delta) - fx(x)) / delta
fx_dx_centered = lambda x, delta: (fx(x + delta) - fx(x - delta)) / (2 * delta)
fx_dx_delayed = lambda x, delta: (fx(x) - fx(x - delta)) / delta

fx_2dx = lambda x, delta: (fx(x + delta) - 2 * fx(x) + fx(x - delta)) / delta ** 2
```

```python
delta_x = 0.01

plt.plot(x, fx(x), color = "cyan")
plt.plot(x, fx_dx(x), color = "black")
plt.plot(x, fx_dx_delayed(x, delta_x), color = "red")
plt.plot(x, fx_dx_centered(x, delta_x), color = "green")
plt.plot(x, fx_dx_forward(x, delta_x), color = "blue")
# plt.plot(x, fx_2dx(x, delta_x), color = "orange")
plt.show()
```
