```python
from random import randint 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-4, 4, 100)
y = np.exp(x)
```

```python
plt.plot(x, y)
```

```python
def exp2(x, order):
    return np.sum([(x**n) / np.math.factorial(n) for n in range(order+1)])
```

```python
plt.plot(x, y, color='#000000')
colors = []
labels = []
for order in range(2, 11, 1):
    colors.append('#%06X' % randint(120, 0xEEEEEE))
    labels.append(f'Orden {order}')
    plt.plot(x, [exp2(a, order) for a in x], color=colors[order - 2])
plt.scatter([0], [1], color="red")
plt.legend(labels)
```
