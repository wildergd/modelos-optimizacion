```python
from random import randint
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
def sin_fn(x, order):
    return np.sum([((-1) ** n) * (x**(2 *n + 1)) / np.math.factorial(2 * n + 1) for n in range(order+1)])
```

```python
plt.plot(x, y, color='#000000')
colors = []
labels = []
for order in range(2, 11, 1):
    colors.append('#%06X' % randint(120, 0xEEEEEE))
    labels.append(f'Orden {order}')
    plt.plot(x, [sin_fn(a, order) for a in x], color=colors[order - 2])
plt.scatter([0], [0], color="red")
plt.legend(labels)
```
